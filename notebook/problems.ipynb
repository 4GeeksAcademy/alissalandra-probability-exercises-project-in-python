## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import numpy as np
import random

sample_space = [1,2,3,4,5,6]
number_of_trials = 1000

results_1 = np.random.choice(sample_space, size=(1000))
results_2 = np.random.choice(sample_space, size=(1000))

total_of_results = results_1 + results_2

greater_than_7_or_even = np.sum((total_of_results > 7) | (total_of_results %2 == 0))

probability = greater_than_7_or_even / number_of_trials

print(probability)
print (f"The probability of getting greater than 7 or an even number is {probability}")
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

          
# TODO
#generate 5 random ball picks

def select_five_balls():
    pulled_balls = []
    for i in range(5):
        random_ball_index = np.random.randint(0, 60)
        random_ball_color = ball_box[random_ball_index]
        pulled_balls.append(random_ball_color)

    pulled_balls = np.array(pulled_balls)
    return pulled_balls

#run 1000 tests

def run_10_tests():
    event1_count = 0  # Take 3 white and 2 red
    event2_count = 0  # Take all of the same color
   
    for i in range(1000):
        drawnball = select_five_balls()

        white_count = np.count_nonzero(drawnball == 'White')
        red_count = np.count_nonzero(drawnball == 'Red')
        green_count = np.count_nonzero(drawnball == 'Green')

        if white_count == 3 and red_count == 2:
            event1_count += 1

        if white_count == 5 or red_count == 5 or green_count == 5:
            event2_count += 1

    probability_event1 = event1_count / 1000
    probability_event2 = event2_count / 1000
        
    return probability_event1, probability_event2

prob_event1, prob_event2 = run_10_tests()

print("Probability of taking 3 white and 2 red:", prob_event1)
print("Probability of taking all of the same color:", prob_event2)

```
